### sentiments of reviews 

```python
def sentiment_r(l):
  if l<3:
    return 'negative'
  elif l== 3:
    return 'neutral'
  else:
    return 'positive'
```

```python
df_gp['sentiment'] = df_gp['score'].apply(sentiment_r)
df_gp
```

```python
plt.pie(df_gp['sentiment'].value_counts(),labels = df_gp['sentiment'].value_counts().index, autopct='%1.1f%%',colors=['yellowgreen', 'orangered', 'gold'])
plt.title("Sentiment of reviews on Google play store")
plt.show()
```

```python
df_as['sentiment'] = df_as['score'].apply(sentiment_r)
df_as
```

```python
plt.pie(df_as['sentiment'].value_counts(),labels = df_as['sentiment'].value_counts().index, autopct='%1.1f%%',colors=['orangered','yellowgreen', 'gold'])
plt.title("Sentiment of reviews on Apple app store")
plt.show()
```

### Data cleaning

```python
from wordcloud import WordCloud ,STOPWORDS
```

```python
dfn_gp = df_gp[df_gp['sentiment'].isin(['negative','neutral'])] 
dfp_gp = df_gp[df_gp['sentiment']=='positive']
```

```python
dfn_gp = dfn_gp[['content']]
dfn_gp.dropna(inplace = True)
dfp_gp = dfp_gp[['content']]
dfp_gp.dropna(inplace = True)
```

```python
stw = list(STOPWORDS) + ['good','nice','worst','bad','great','loco']
```

```python
nl = dfn_gp['content'].tolist()
pl = dfp_gp['content'].tolist()
```

```python
neg_text = ' '.join(nl).lower()
pos_text = ' '.join(pl).lower()
```

```python
!pip3 install yake
```

```python
import yake
```

```python
kw_extractor = yake.KeywordExtractor(lan = 'en', n = 3, dedupLim = 0.8 ,dedupFunc= 'seqm', windowsSize = 1,top = 50)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
sw_nltk = stopwords.words('english')
```

```python
sw_nltk+= ['good','bad','worst','worst','app','loco','waste','time','game','play','hai']
```

```python
words = [word for word in neg_text.split() if word.lower() not in sw_nltk]
new_text = " ".join(words)
```

```python
keywords = kw_extractor.extract_keywords(new_text)
keywords
```

```python
l = dict([(x[0],1/x[1]) for x in keywords])
```

```python
li = [x[0] for x in keywords]
```

```python
wc = WordCloud(background_color="white",colormap = sns.color_palette("dark:salmon_r", as_cmap=True),stopwords=stw,collocations = True).generate_from_frequencies(l)
plt.title("Keywords from negative reviews")
plt.imshow(wc)
```

```python
wc.to_file('/content/drive/MyDrive/ML/app analysis/gp_neg.png')
```

```python
sw = stopwords.words('english')
sw+=['good','bad','worst','worst','app','loco','waste','time','game','play','hai','nice','live','amazing','application','awesome']
words = [word for word in pos_text.split() if word.lower() not in sw]
ntext = " ".join(words)
kw = kw_extractor.extract_keywords(ntext)
kw
```

```python
p = dict([(x[0],1/x[1]) for x in kw])
```

```python
wc = WordCloud(background_color="white",colormap = sns.cubehelix_palette(start=.5, rot=-.75, as_cmap=True),stopwords=stw,collocations = True).generate_from_frequencies(p)
plt.title("Keywords from positive reviews")
plt.imshow(wc)
```
